```python
import torch
import torch.nn as nn
```

```python
# Definición del modelo

class InceptionNaives(nn.Module):
  def __init__(self, channel, outputs):
    super(InceptionNaives, self).__init__()
    self.channel = channel
    self.outputs = outputs

    self.net_architecture()

  def net_architecture(self):
    # Convolución 1x1
    self.Conv1x1 = nn.Conv2d(self.channel, self.outputs[0], kernel_size = 1, stride = 1, padding = 0)

    # Convolución 1x1 - 3x3
    self.Branch3x3 = nn.Sequential(
        nn.Conv2d(self.channel, self.outputs[1], kernel_size = 1, stride = 1, padding = 0),
        nn.Conv2d(self.outputs[1], self.outputs[2], kernel_size = 3, stride = 1, padding = 1)
    )
    # Convolución 1x1 - 5x5
    self.Branch5x5 = nn.Sequential(
        nn.Conv2d(self.channel, self.outputs[3], kernel_size = 1, stride = 1, padding = 0),
        nn.Conv2d(self.outputs[3], self.outputs[4], kernel_size = 5, stride = 1, padding = 2)
    )
    # Max pooling - 1x1
    self.maxpool = nn.Sequential(
        nn.MaxPool2d(kernel_size = 3, stride = 1, padding = 1),
        nn.Conv2d(self.channel, self.outputs[5], kernel_size = 1, stride = 1, padding = 0)
    )

  def forward(self, x_data):
    return torch.cat([self.Conv1x1(x_data), self.Branch3x3(x_data), self.Branch5x5(x_data), self.maxpool(x_data)], dim = 1)

'''
Nota #1: La salida del anterior es la entrada del siguiente cuando usas snn.Sequential
Nota #2: dim 1 sirve para hacer un flatten de tensores, en lugar de crear 1 por cada operación
'''
```

```python
# Definición de parametros

channel = 192
outputs = [64, 96, 128, 16, 32, 32]
size = [28, 28]
```

```python
# Generación de los datos

data = torch.rand(1, channel, size[0], size[1])
```

```python
# Instanciación del modelo

model = InceptionNaives(channel, outputs)
```

```python
# Ejecucuón del modelo

prediction = model(data)
print(prediction.shape)
```
